### 4.7 Example: Random Walks

```python
import random
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:100])
```

```python
nsteps = 1000
rng = np.random.default_rng(seed=12345)
```

```python
draws = rng.integers(0, 2, size = nsteps)
```

```python
steps = np.where(draws == 0, 1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
plt.plot(walk)
```

```python
# Find the first moment the person is away from the starting point by 10 steps
(np.abs(walk) >= 20).argmax()
```

```python
plt.plot(walk[:250])
```

#### Simulating Many Walks

```python
nwalks = 5000
nsteps = 1000
draws = rng.integers(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(axis = 1)
```

```python
walks
```

```python
walks.max()
```

```python
hits30 = (np.abs(walks) >= 30).any(axis=1)
hits30
```

```python
hits30.sum()
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis=1)
crossing_times
```

```python
crossing_times.mean()
```
